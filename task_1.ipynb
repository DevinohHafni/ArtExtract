```python
from datasets import load_dataset

dataset = load_dataset("huggan/wikiart", split="train[:2000]")
```

```python
from torchvision import transforms

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

def transform_example(example):
    image = example["image"]

    # FIX: ensure it's PIL
    if isinstance(image, list):
        image = image[0]

    example["image"] = transform(image)
    return example

dataset = dataset.map(transform_example)
```

```python
print(dataset[0])
```

```python
from sklearn.preprocessing import LabelEncoder

labels = [item['style'] for item in dataset]

le = LabelEncoder()
encoded_labels = le.fit_transform(labels)
```

```python
def add_label(example, idx):
    example["label"] = int(encoded_labels[idx])
    return example

dataset = dataset.map(add_label, with_indices=True)
```

```python
dataset = dataset.train_test_split(test_size=0.2)

train_data = dataset['train']
test_data = dataset['test']
```

```python
import torch
from torch.utils.data import DataLoader

def collate_fn(batch):
    images = []
    labels = []

    for item in batch:
        img = item["image"]

        if isinstance(img, list):
            img = img[0]

        images.append(img)
        labels.append(item["label"])

    return {
        "image": torch.stack(images),
        "label": torch.tensor(labels)
    }

train_loader = DataLoader(train_data, batch_size=16, shuffle=True, collate_fn=collate_fn)
test_loader = DataLoader(test_data, batch_size=16, collate_fn=collate_fn)
```

```python
import torch.nn as nn
import torchvision.models as models

device = "cuda" if torch.cuda.is_available() else "cpu"

model = models.resnet18(pretrained=True)
model.fc = nn.Linear(model.fc.in_features, len(le.classes_))
model = model.to(device)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
```

```python
def train(model, loader):
    model.train()
    total_loss = 0

    for batch in loader:
        images = batch["image"].to(device)
        labels = batch["label"].to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(loader)
```

```python
from torchvision import transforms

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

def collate_fn(batch):
    images = []
    labels = []

    for item in batch:
        img = item["image"]

        # FIX 1: unwrap list
        if isinstance(img, list):
            img = img[0]

        # FIX 2: convert to tensor if not already
        if not isinstance(img, torch.Tensor):
            img = transform(img)

        images.append(img)
        labels.append(item["label"])

    return {
        "image": torch.stack(images),
        "label": torch.tensor(labels)
    }
```
